```python
from skmlm import NN_MLM
from skmlm import FCM_MLM
from skmlm import MLM
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.datasets import load_iris,load_wine,load_boston,load_breast_cancer
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = load_iris()
wine = load_wine()
boston = load_boston()
breast_cancer = load_breast_cancer()
```

```python
def test_MLM(df):
    size = len(df.target)
    scores_mean = []
    for k in np.arange(20,size-40,10):
        clf = make_pipeline(MinMaxScaler(),NN_MLM(rp_number=k))
        scores = cross_val_score(clf, df.data,df.target, cv=10,scoring='accuracy')
        scores_mean.append(scores.mean())
    return scores_mean
```

```python
res_wine = test_MLM(wine)
res_iris = test_MLM(iris)
res_boston = test_MLM(boston)
res_breast_cancer = test_MLM(breast_cancer)
```

```python
print("Dataset Iris: ",np.max(res_iris))
print("Dataset Wine: ",np.max(res_wine))
print("Dataset Boston: ",res_boston)
print("Dataset Breast_cancer",np.max(res_breast_cancer[:-1]))
```
